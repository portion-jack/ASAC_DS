```python
x = np.linspace(start=0, stop=10, num=11)
y = x**2

print(x)
print(y)
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.1,0.5,0.4,0.3])


# Main_graph
axes1.plot(x,y,'r')
axes1.set_xlabel("x")
axes1.set_ylabel("y")
axes1.set_title("MainGraph")

# sub graph (in Main_graph)
axes2.plot(y,x,'g')
axes2.set_xlabel("y")
axes2.set_ylabel("x")
axes2.set_title("SubGraph")
```

```python
# 그래프 위치를 지정하지 않고 분할해서 그리는 방법
fig, axes = plt.subplots()

axes.plot(x,y,'r')
axes.set_title('sub1')
```

```python
fig, axes = plt.subplots(nrows=1,ncols=3)
i = 0
for ax in axes:
    ax.plot(x,y)
fig.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=2,ncols=2)
# 0,0
axes[0,0].plot(x,x**2,x,x**2+100)
axes[0,0].set_title('axes 0 ,0')

# 0,1
axes[0,1].plot(x,x**2,x,x**3)
axes[0,1].set_title('axes 0,1')

# 1,0
axes[1,0].plot(x,x**2,x,x**2+1000)
axes[1,0].set_title('axes 1,0')

# 1,1
axes[1,1].plot(x,x**2,x,x**2+100)
axes[1,1].set_yscale("log")
axes[1,1].set_title('axes 1,1')

fig.tight_layout()
```
