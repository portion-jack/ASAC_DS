```python
## 
baseball_player = pd.DataFrame(
    {
        'player_name':['권혁','이범호','필','박병호','양준혁'],
        'team_no': [1993,2001,2001,2010,None]
    }
)
baseball_player
```

```python
baseball_player_team = pd.DataFrame(
    {
        'team_name' : ['한화','두산','기아','넥센'],
        'team_home' : ['대전','서울','광주','고척'],
        'team_id'   : [1993, 1999, 2001, 2010]
    }
)
baseball_player_team
```

```python
baseball_player
```

```python
# merge
## column 기준 on index 기준 index
# default : how="inner"
# pd.merge(left_df,right_df,left_on = "col_1",right_on='col_2', how="inner")
# pd.merge(left_df,right_df,left_on = "col_1",right_on='col_2', how="outer")

pd.merge(baseball_player,baseball_player_team,
         left_on='team_no',right_on='team_id',
         how='inner')
```

```python
pd.merge(baseball_player,baseball_player_team,
         left_on='team_no',right_on='team_id',
         how='outer')
```

```python
# 혹시 2개의 df의 기준이 모두 컬럼이며 같을 때
temp=baseball_player.rename(columns={'team_no':'team_id'})

pd.merge(temp,baseball_player_team,
         on='team_id')
```

```python
# 팀에 대한 정보가 없는 선수
pd.merge(baseball_player,baseball_player_team,
         left_on='team_no',right_on='team_id',
         how='outer')
```

```python
pd.merge(baseball_player,baseball_player_team,
         left_on='team_no',right_on='team_id',
         how='outer')
```

```python
temp=pd.merge(baseball_player,baseball_player_team,
         left_on='team_no',right_on='team_id',
         how='outer')

temp[temp['player_name'].isna()]
```

```python
temp=pd.merge(baseball_player,baseball_player_team,
         left_on='team_no',right_on='team_id',
         how='outer')

temp[(temp['player_name'].isna()) | (temp['team_id'].isna())]
```
