```python
from __private__ import my_api_key
import requests
import pandas as pd
from bs4 import BeautifulSoup

params = {
    "key": my_api_key(),
    "itemPerPage":"50",
}

base_url = "http://kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.xml"

res = requests.get(base_url,params=params)
if res.ok:
    result=res.content.decode('utf-8')
    soup = BeautifulSoup(result,"xml")
else:
    raise ValueError

result_li=soup.find_all("movieListResult")[0].find_all("movieList")[0].find_all('movie')
```

```python
using_columns = [
    "movieCd",
    "movieNm",
    "movieNmEn",
    "openDt",
    "typeNm",
    "directors"
    ]
```

```python
import numpy as np

_df=pd.DataFrame(columns=using_columns,data=np.zeros((1,len(using_columns))))
df = pd.DataFrame(columns=using_columns)
```

```python
for i in range(len(result_li)):
    for column in using_columns:
        _df.at[0,column]=result_li[i].find(column).get_text() 
    df = pd.concat((df,_df))
```

```python
df['year']=df['openDt'].apply(lambda x : x[:4])
df['month']=df['openDt'].apply(lambda x : x[4:6])
df['date']=df['openDt'].apply(lambda x : x[6:])
df.set_index('movieCd',inplace=True)

df
```
