```python
from __private__ import my_api_key
import requests
import pandas as pd

params = {
    "key": my_api_key(),
    "itemPerPage":"50",
}

base_url = "http://kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.json"

response = requests.get(base_url,params=params)
if response.ok:
    result = response.json()
else:
    raise ValueError

result_li = result['movieListResult']['movieList']
```

```python
df=pd.DataFrame(result_li)
```

```python
def my_director_try(x):
    try: 
        return x[0]['peopleNm']
    except :
        return ""

def my_company_try(x):
    try: 
        return x[0]['companyNm']
    except :
        return ""
    
df['directors']=df['directors'].apply(lambda x: my_director_try(x))
df['companys']=df['companys'].apply(lambda x: my_company_try(x))
```

```python
df['year']=df['openDt'].apply(lambda x : x[:4])
df['month']=df['openDt'].apply(lambda x : x[4:6])
df['date']=df['openDt'].apply(lambda x : x[6:])
df.set_index('movieCd', inplace=True)

df
```

```python
base_url = "http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieInfo.json"

for movieCd in df.index:
    params = {
        "key" :my_api_key(),
        "movieCd": movieCd,
    }

    response = requests.get(base_url,params=params)
    
    if response.ok:
        result = response.json()
    
    result_dict=result['movieInfoResult']['movieInfo']
    
    if result_dict["genres"] == []:
        df.at[movieCd,'genre'] = ''
    elif result_dict["genres"] != []:
        df.at[movieCd,'genre'] = result_dict['genres'][0]['genreNm']
    if result_dict["actors"] == []:
        df.at[movieCd,'actor_num'] = ''
        df.at[movieCd,'main_actor_name'] = ''
        df.at[movieCd,'main_actor_role'] = ''
    elif result_dict["actors"] != []:
        df.at[movieCd,'actor_num'] = int(len(result_dict['actors']))
        df.at[movieCd,'main_actor_name'] = result_dict['actors'][0]['peopleNm']
        df.at[movieCd,'main_actor_role'] = result_dict['actors'][0]['cast']
```
