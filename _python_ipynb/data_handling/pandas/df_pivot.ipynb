```python
import pandas as pd
```

```python
path = 'data/11_sales-funnel.xlsx'
data = pd.read_excel(path,sheet_name='Sheet1')
```

```python
data.head()
```

```python
pd.pivot_table(data=data, index=['Name'])
```

```python
pd.pivot_table(data=data, index=['Name','Manager'])
```

```python
_df=pd.pivot_table(data=data, index=['Manager','Rep'])
_df.loc['Debra Henley']
```

```python
pd.pivot_table(data=data, index=['Manager','Rep'],
                   values=['Price'])
```

```python
import numpy as np

pd.pivot_table(data,index=['Manager','Rep'],
               values=['Price'],
               aggfunc=[np.sum])
```

```python
pd.pivot_table(data,index=['Manager','Rep'],
               values=['Price','Quantity'],
               aggfunc=[np.sum,len])
```

```python
pd.pivot_table(data,index=['Manager','Rep'],
               values=['Price'],
               columns=['Product'],
               aggfunc=[np.sum])
```

```python
pd.pivot_table(data,index=['Manager','Rep'],
               values=['Price'],
               columns=['Product'],
               aggfunc=[np.sum],
               fill_value=0)
```

```python
pd.pivot_table(data,index=['Manager','Rep'],
               values=['Price','Quantity'],
               columns=['Product'],
               aggfunc=[np.sum],
               fill_value=0,
               margins=True)
```

```python
# pivot table

## 가로 : index
## 세로 : values, columns(항목별)
```

```python
table=pd.pivot_table(data,index=['Manager','Rep'],
               values=['Price','Quantity'],
               columns=['Product'],
               aggfunc={'Price':[np.sum,np.mean],
                        'Quantity':[len]},
               fill_value=0)
```

```python
table.query("Manager=='Debra Henley'")
```
